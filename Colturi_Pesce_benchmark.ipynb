```python
# import to generate random data
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.distance import pdist, squareform as sf
```

# Gaussian

```python
# generate 2D gaussian clusters

# means
mean1=[1,0]
mean2=[-2,1]
mean3=[-5,-2]
mean4=[3,-5]

# covariances
cov1=[[0.5,0],[0,0.2]]
cov2=[[0.5,0],[0,0.2]]
cov3=[[0.3,0],[0,0.6]]
cov4=[[0.5,0],[0,0.3]]

# sample gaussian
x1, y1 = np.random.multivariate_normal(mean1, cov1, 500).T
x2, y2 = np.random.multivariate_normal(mean2, cov2, 100).T
x3, y3 = np.random.multivariate_normal(mean3, cov3, 200).T
x4, y4 = np.random.multivariate_normal(mean4, cov4, 500).T

x = np.concatenate([x1,x2,x3,x4])
y = np.concatenate([y1,y2,y3,y4])
# labels are the clustering
label = np.concatenate([np.repeat(0,x1.shape[0]), np.repeat(1,x2.shape[0]), np.repeat(2,x3.shape[0]), np.repeat(3,x4.shape[0])])

X = np.column_stack([x,y])

plt.scatter(X[:,0],X[:,1], c=label, s=5)
```

```python
# distance metrix, euclidean
DM = pdist(X)
plt.figure()
plt.hist(DM, bins=200)
plt.show()
```

# Uniform

```python
# uniform data
X = np.random.uniform(-1,1,(800,2))
plt.scatter(X[:,0], X[:,1], s=5)
plt.show()
```

```python
# euclidean distance for uniform data
DM = pdist(X)
plt.figure()
plt.hist(DM, bins=100)
plt.show()
```
